```python
plt.figure(figsize=(8,5))
ax = sns.countplot(x='Age Group', data=df,palette=['red','blue'],hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

Totle Sales by State

```python
State_by_sale = df.groupby(['State','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
plt.Figure(figsize=(20,20))
ax= sns.barplot(x='Amount',y='State',data=State_by_sale,hue='Gender')
plt.xticks(rotation=40)
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
for bars in ax.containers:
    ax.bar_label(bars,fmt="%.0f")
plt.show()
```

from above graphs we can see most of the buyers in every states are females and even the purchasing power of females are greater then the men

```python
df["Marital_Status"]= df['Marital_Status'].replace({0:"Married",1:"Unmarried"})
marital_status=df.groupby('Marital_Status',as_index=False)['User_ID'].nunique()
plt.pie(x=marital_status["User_ID"],labels=marital_status['Marital_Status'],autopct='%1.2f%%')
plt.show()
```

from above chart we can see married people more.

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
sale_age= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sale_age,palette="Set2", hue='Age Group')
plt.show()
```

```python
df.columns
```

```python
order_by_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(data=order_by_state,y="State",x="Orders",hue='State')
plt.xticks(rotation=30)
plt.show()
```

TOP 10 State acc. to Sales

```python

Top_state_sale=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
less_state_sale=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).tail(10)
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(data=Top_state_sale, x='Amount',y="State",hue='State', ax=axes[0])
sns.barplot(data=less_state_sale, x='Amount',y="State",hue='State', ax=axes[1])
axes[0].set_title("TOP 10 STATE BY SALE")
axes[1].set_title("BOTTOM 10 STATE BY SALE")
plt.show()
```

```python
Status_by_sales= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(data=Status_by_sales,x='Marital_Status',y='Amount',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars,fmt="%.0f")
    plt.show()
```

From above graphs we can see the most of buyres are married women and they are high purchasing power.

```python
plt.Figure(figsize=(15,13))
ax = sns.countplot(data=df, x='Occupation',hue='Occupation')
plt.xticks(rotation=45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.grid(True,axis='y')
plt.show()

```

```python
occuption_sales= df.groupby("Occupation",as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
plt.Figure(figsize=(20,7))
sns.barplot(data=occuption_sales,x="Occupation",y='Amount',hue='Occupation')
plt.show()
```

From above you can see most of the buyres are IT Sectors ,Healthcare and Aviation.
The most of less buyres are Agriculture, Textile and Construction.

```python
plt.Figure(figsize=(20,15))
ax=sns.countplot(data=df,x='Product_Category',hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=15)
plt.show()
```

```python
plt.Figure(figsize=(20,10))
order_value = df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head()
order_value_L = df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).tail()
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
ax=sns.barplot(data=order_value,x='Product_Category',y='Amount',hue='Product_Category',ax=axes[0])
ax1=sns.barplot(data=order_value_L,x='Product_Category',y='Amount',hue='Product_Category',ax=axes[1])
for bars in ax.containers:
    ax.bar_label(bars,fmt="%0.f")

for bars in ax1.containers:
    ax1.bar_label(bars,fmt="%0.f")
axes[0].set_title("TOP 5 Product_Category BY SALE")
axes[1].set_title("BOTTOM 5 Product_Category  BY SALE")
plt.show()
```

From above graphs you can see most of sold product is Food and Clothing&Appetiger.
The least sold product are office and Veterinary.
